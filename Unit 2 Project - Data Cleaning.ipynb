### Step 5: Summarize Actions Taken & Justify Approach

######################### TO DO ###################################

####### FOLLOW PROJECT DIRECTIONS TO COMPLETE THIS STEP ###########

####### TO DO ###### - 
the zscore between 3 and -3 keeps pulling up hundreds of properties for each column so in order to keep as much relevant data as possible I have rarely used this criteria to get rid of information.

DATA:
Bedrooms - although there are studio apartments which do not include a bedroom, the sq footage of the 0 bedrooms did not match on most of them, so I deleted all zscores less than -3 and also the zscores greater than 4.  There could be houses with more rooms, but the higher ones looked more like typos when comparing rooms to bathrooms and square footage.

Bathrooms - it is not uncommon for a studio apartment to have .75 bath.  So I made sure all low bathrooms also had a low sq footage. It's also plausible that a house with 10k sq ft+ would have 7-8 bathrooms. so I only removed the bathrooms higher than 7 and lower than .5 because those did not make sense with the square footage or livability.

Sqft Living - mansions aren't as much of an outlier as they used to be so I increased the zscore to delete to 6 so it would include the more common larger homes.

sqft Lot - got rid of zscores higher than 6 and less than -3.  guess the apartments and farm properties are gone. I don't like deleting with the zscore. Especially in a database that has such a wide range of criteria...I'm sure I'm getting rid of the most amazing or historical properties.

Sqft Above - only got rid of zscore greater than 5.  Compared to sqft living, sqft basement and sqft above to make sure things made sense. 

Sqft Basement - many 2k basements were in the zscore range of 3.  That is not an unreasonable size for a basement, expecially when the above sq footage is the same size.  I only deleted basement zscores greater than 5.  there were none less than -3

Year Built - deleted zscores greater than 2 and less than -3.  This got rid of an outlier(1000) and houses not built yet.

Sqft Lot 15 - removed definite typo value with zscore greater than 4

SALES:
Price - removed the top 3 prices since they were quite a bit higher than the rest - this would be much easier to see outliers if we had a price/sqft column
